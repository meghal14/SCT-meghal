```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("/content/Mall_Customers (1).csv")
df.head(2)
```

```python
df["Gender"].value_counts()
```

```python
df1=df.replace({"Female":1, "Male":0})
df1.head(2)
```

```python
sns.pairplot(df)
```

```python
wcss=[]
for i in range(2,10):
  model=KMeans(n_clusters=i,init="k-means++")
  model.fit(df1)
  wcss.append(model.inertia_)
```

```python
plt.plot([ i for i in range(2,10)],wcss)
plt.xlabel("no of cluster")
plt.ylabel("wcss")
plt.grid(axis="x")
```

```python
model = KMeans(n_clusters=5)
df1["prediction"] = model.fit_predict(df1)
sns.pairplot(df1, hue="prediction")
plt.show()
```
